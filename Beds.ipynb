# Future Prediction of upcoming 12 months

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
# Preprocess the data
df2['Date'] = pd.to_datetime(df2[['Year', 'Month']].assign(day=1))
df2 = df2.drop(columns=['Year', 'Month'])

# Aggregate the data to find the top 3 beds in terms of usage
top_beds = df2.groupby(['WARD', 'BED_TYPE', 'BED_NUMBER']).size().reset_index(name='counts')
top_beds = top_beds.nlargest(3, 'counts')
```

```python
# Filter the data for the top 3 beds
top_beds_data = df2[df2[['WARD', 'BED_TYPE', 'BED_NUMBER']].apply(tuple, axis=1).isin(
    top_beds[['WARD', 'BED_TYPE', 'BED_NUMBER']].apply(tuple, axis=1)
)]
```

```python
# Function to create time series data for each bed
def create_time_series(df2, ward, bed_type, bed_number):
    bed_data = df2[(df2['WARD'] == ward) & (df2['BED_TYPE'] == bed_type) & (df2['BED_NUMBER'] == bed_number)]
    bed_data = bed_data.groupby('Date').size().reset_index(name='counts')
    bed_data = bed_data.set_index('Date').asfreq('MS').fillna(0)
    return bed_data['counts']
```

```python
# Prepare time series data for each top bed
time_series_data = {}
for index, row in top_beds.iterrows():
    ward, bed_type, bed_number = row['WARD'], row['BED_TYPE'], row['BED_NUMBER']
    time_series_data[(ward, bed_type, bed_number)] = create_time_series(top_beds_data, ward, bed_type, bed_number)
```

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing

# Function to forecast future usage using ETS
def forecast_ets(time_series, steps=12):
    if len(time_series) < 12:
        print(f"Warning: Insufficient data ({len(time_series)} months) to estimate seasonality. Using Simple Exponential Smoothing instead.")
        model = SimpleExpSmoothing(time_series)
    else:
        seasonal_periods = min(12, len(time_series))  # Adjust seasonal periods if necessary
        model = ExponentialSmoothing(time_series, seasonal='add', seasonal_periods=seasonal_periods)
    
    model_fit = model.fit()
    forecast = model_fit.forecast(steps)
    return forecast
```

```python
# Function to plot the original and forecasted data
def plot_forecast(time_series, forecast, title):
    # Create date range for forecasted period
    forecast_dates = pd.date_range(start=time_series.index[-1] + pd.DateOffset(months=1), periods=len(forecast), freq='MS')
    
    plt.figure(figsize=(10, 6))
    plt.plot(time_series.index, time_series, label='Original')
    plt.plot(forecast_dates, forecast, label='Forecast', color='red')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Usage')
    plt.legend()
    plt.show()

# Forecast and plot for each top bed
for bed, ts_data in time_series_data.items():
    forecast = forecast_ets(ts_data)
    plot_forecast(ts_data, forecast, title=f'Forecast for {bed}')
```
